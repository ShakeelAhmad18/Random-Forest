```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import train_test_split
from matplotlib import pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
```

```python
full_data=pd.read_csv('diabetes.csv')
```

```python
full_data
```

```python
x=full_data.drop('Outcome',axis=1)
y=full_data['Outcome']
```

```python
x
```

```python
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=20)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
model=RandomForestClassifier()
```

```python
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
y_pred
```

```python
print(y_test)
```

```python
model.predict([[1, 85, 66, 29, 0,26.6 ,0.351, 31]])
```

```python
accuracy_score(y_test,y_pred)
```

```python
cm=confusion_matrix(y_test,y_pred)
cm
```

```python
plt.figure(figsize=(5,6))
g=sns.heatmap(cm,cmap='Greens',annot=True)
g.set_xticklabels(labels=['No Diabetes (0)' , 'Diabetes (1)'], rotation=30)
g.set_yticklabels(labels=['No Diabetes (0)' , 'Diabetes (1)'], rotation=0)
plt.ylabel('True Label', fontsize=14)
plt.xlabel('Predicted Label', fontsize=14)
plt.title('Confusion Matrix', fontsize=16)
plt.show()
```

## Create the model that can classify heart disease of the patients using Random Forest
## Classifier, Predict heart disease for the follwing patient parameters, generate the classification
## report of the model, draw confusion matrix.


```python
data=pd.read_csv('D:\Machine Learning\heart.csv')
data
```

```python
x=data.drop('HeartDisease',axis=1)
y=data['HeartDisease']
```

```python
le_Sex = LabelEncoder()
le_ChestPainType = LabelEncoder()
le_RestingECG = LabelEncoder()
le_ExerciseAngina = LabelEncoder()
le_ST_Slope = LabelEncoder()
x['Sex_n'] = le_Sex.fit_transform(x['Sex'])
x['ChestPainType_n'] = le_ChestPainType.fit_transform(x['ChestPainType'])
x['RestingECG_n'] = le_RestingECG.fit_transform(x['RestingECG'])
x['ExerciseAngina_n'] = le_ExerciseAngina.fit_transform(x['ExerciseAngina'])
x['ST_Slope_n'] = le_ST_Slope.fit_transform(x['ST_Slope'])
```

```python
x
```

```python
X = x.drop(['Sex','ChestPainType','RestingECG','ExerciseAngina','ST_Slope'],axis=1)
```

```python
X
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.20,random_state=20)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
nmodel=RandomForestClassifier()
```

```python
nmodel.fit(x_train,y_train)
```

```python
y_pre=nmodel.predict(x_test)
y_pre
```

```python
y_test
```

```python
accuracy_score(y_test,y_pre)
```

```python
nmodel.predict([[37, 130, 283, 0, 98, 0.0, 1, 1, 2, 0, 2]])
```

```python
result=y_test-y_pre
result
```
